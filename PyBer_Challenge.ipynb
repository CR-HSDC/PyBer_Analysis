### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head(100)
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_ride_count_df = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_ride_count_df.head()
```

```python
# 2. Get the total drivers for each city type
total_driver_count_df = city_data_df.groupby(["type"]).sum()["driver_count"]
total_driver_count_df.head()
```

```python
#  3. Get the total amount of fares for each city type
total_fare_count_df = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fare_count_df.head()
```

```python
#  4. Get the average fare per ride for each city type. 
mean_fare_df = pyber_data_df.groupby(["type"]).mean()["fare"]
mean_fare_df.head()
```

```python
# 5. Get the average fare per driver for each city type. 
mean_fare_per_driver_df = total_fare_count_df/total_driver_count_df
mean_fare_per_driver_df.head()
```

```python
#  6. Create a PyBer summary DataFrame. 

pyber_summary_df = pd.DataFrame({
               "Total Rides": total_ride_count_df,
               "Total Drivers": total_driver_count_df,
               "Total Fares": total_fare_count_df,
               "Average Fare per Ride": mean_fare_df,
                "Average Fare per Driver": mean_fare_per_driver_df })

pyber_summary_df.head()
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
```

```python
# 1. Read the merged DataFrame

# STUDENT COMMENT: THIS LINE IS REDUNDANT AND DOES NOT MATCH UP WITH COURSE NOTES FOR MODULE 5 DELIVERABLE 
# DESCRIPTION: CODES BEGINS IN STEP 2, STEPS ARE OFFSET BY 1; MODULE 5  STEP 1 DESCRIPTION PASTED BELOW FOR REFERENCE:

# In Step 1, create a new DataFrame with multiple indices using the groupby() function on the 
# "type" and "date" columns of the pyber_data_df DataFrame, 
# then apply the sum() method on the "fare" column to show the total fare amount for each date.
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
by_date_type_count_df = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]
by_date_type_count_df.tail()
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
by_date_type_count_df = by_date_type_count_df.reset_index()
by_date_type_count_df.tail()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
by_date_type_count_pivot = by_date_type_count_df.pivot(index="date", columns="type",values="fare")
by_date_type_count_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

JanToApr_type_count_pivot = by_date_type_count_pivot.loc[((by_date_type_count_pivot.index>='2019-01-01') & (by_date_type_count_pivot.index<='2019-04-29'))]

JanToApr_type_count_pivot.tail()
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
JanToApr_type_count_pivot.index = pd.to_datetime(JanToApr_type_count_pivot.index)
JanToApr_type_count_pivot.head()
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
JanToApr_type_count_pivot.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
Weekly_Resample_df = JanToApr_type_count_pivot.resample("W").sum()
Weekly_Resample_df.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')


# Create a box-and-whisker plot for the urban cities ride count.
#x_labels = ["Urban"]
fig, ax = plt.subplots(figsize=(15,5))

ax.set_title('Total Fare by City Type')
ax.set_ylabel('Fare ($USD)')
#ax.set_yticks(np.arange(10, 41, step=2.0))
#ax.legend(["Rural","Sub","Urb"]);


ax.plot(Weekly_Resample_df)
ax.legend(["Rural", "Suburban","Urban"],loc='center',title="Type");

import matplotlib.dates as mdates
#ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))

#Format X axid to display months and set margin to zero.
fmt = mdates.DateFormatter('%b')
X = plt.gca().xaxis
locator = mdates.MonthLocator()  # every month
X.set_major_locator(locator)
X.set_major_formatter(fmt)
plt.margins(x=0)
plt.savefig("analysis/PyBer_fare_summary.png")
plt.show()



```
